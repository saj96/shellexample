```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 7))
x_data = ['A', 'B', 'C', 'D', 'E']
y_data = [5, 10, 15, 20, 25]
x_pos = np.arange(len(x_data))
plt.bar(x_pos, y_data)
plt.xticks(x_pos, x_data)
plt.ylabel('Value')
plt.title('Simple Bar Chart')
plt.show()
```

# Scatter age vs survived

```python
import matplotlib.pyplot as plt
data = %sql SELECT hex(survived), age from titanic
# print(data)

age = []
survived = []
for d in data :
    survived.append(d[0])
    age.append(d[1])

print(survived)
plt.scatter(age, survived)
    
```
